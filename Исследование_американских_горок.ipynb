* ``` python df.loc[~df.duplicated(subset=['Coaster_Name','Location','Opening_Date'])]``` Это фильтрует датафрейм df, оставляя только строки, которые не являются дубликатами. Условие проверяет на дубликаты по подмножеству столбцов ['Coaster_Name', 'Location', 'Opening_Date']. Знак ~ перед df.duplicated(...) инвертирует результат, сохраняя только уникальные строки.
* .reset_index(drop=True) - Этот метод сбрасывает индексы датафрейма и сгенерирует новые индексы. Аргумент drop=True указывает, что старые индексы будут удалены, а не сохранены в датафреймах как столбец.
* .copy() - Создает полную копию отфильтрованного датафрейма. Это полезно, чтобы избежать предупреждений о том, что вы работаете с представлениями датафрейма, а не с самим датафреймом.

````python
# Результатом выполнения данной строки кода будет новый датафрейм, содержащий только дублированные строки по указанным столбцам.
df = df.loc[~df.duplicated(subset=['Coaster_Name','Location','Opening_Date'])] \
    .reset_index(drop=True).copy()
````

* В каких местах самые быстрые американские горки (минимум 10 горак)?

````python
df.head()
````

````python
# вы создаёте новый DataFrame, который не содержит строк со значением 'Other' в столбце 'Location'.
df.query('Location != "Other" ')
````

```python
# мы создаем новый DataFrame, который не содержит строк со значением 'Other' в столбце 'Location'.
ax = df.query('Location != "Other"') \
# Групируем по Location
    .groupby('Location')['Speed_mph'] \
# Применям фильтр 'mean','count' и получаем среднию скорость подсчет
    .agg(['mean','count']) \
# Выесняем в каких случаях это число >= 10 (Просто для того чтобы сделать график немного чище)
    .query('count >= 10') \
# Сортируем результат по средней скорости
    .sort_values('mean')['mean'] \
    .plot(kind='barh', figsize=(12, 5), title='Average Coast Speed by Location')

# Устанавливаем подпись оси X
ax.set_xlabel('Average Coaster Speed')
plt.show()
```

````python
ax = df.query('Location != "Other"') \
    .groupby('Location')['Speed_mph'] \
    .agg(['mean','count']) \
    .query('count >= 10') \
    .sort_values('mean')['mean'] \
    .plot(kind='barh', figsize=(12, 5), title='Средняя скорость амереканских горок в зависимости от местоположения')


ax.set_xlabel('Средняя скорость американских горок')
plt.show()
````
